```python
# Applying the log1p transformation to the target variable
y_transformed = np.log1p(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y_transformed,test_size=0.2,random_state=42)
```

```python
x.columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['rating', 'processor_speed','num_cores', 'battery_capacity', 'fast_charging','ram_capacity','internal_memory','screen_size','refresh_rate','num_rear_cameras','primary_camera_rear','primary_camera_front','extended_memory_available','extended_upto']),
        ('cat', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1,min_frequency=3), ['has_5g', 'has_nfc', 'has_ir_blaster']),  # Assuming these are boolean features
        ('cat1', OneHotEncoder(drop='first', sparse_output=False,handle_unknown='ignore',min_frequency=3), ['brand_name', 'processor_brand', 'os', 'resolution'])
    ],
    remainder='passthrough'
)
```

```python
def scorer(model_name, model):

    output = []

    output.append(model_name)

    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])

    # Fit the pipeline
    pipeline.fit(x, y_transformed)

    # K-fold cross-validation
    kfold = KFold(n_splits=10, shuffle=True, random_state=42)
    scores = cross_val_score(pipeline, x, y_transformed, cv=kfold, scoring='r2')

    output.append(scores)

    y_pred = pipeline.predict(x_test)

    output.append(mean_absolute_error(np.expm1(y_test), y_pred))

    return output
```

```python
model_dict = {
    'linear_reg':LinearRegression(),
    'svr':SVR(),
    'ridge':Ridge(),
    'LASSO':Lasso(),
    'decision tree': DecisionTreeRegressor(),
    'random forest':RandomForestRegressor(),
    'extra trees': ExtraTreesRegressor(),
    'gradient boosting': GradientBoostingRegressor(),
    'adaboost': AdaBoostRegressor(),
    'mlp': MLPRegressor(),
    'xgboost':XGBRegressor()
}
```

```python
model_output = []
for model_name,model in model_dict.items():
    model_output.append(scorer(model_name, model))
```

```python
model_df = pd.DataFrame(model_output, columns=['name','r2','mae'])
```
